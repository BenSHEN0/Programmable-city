```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
Adelaide_file_2008=pd.read_csv('IDCJAC0010_023090_2008_Data.csv')
my_data=Adelaide_file_2008
my_data.head()
```

```python
Adelaide_file_2018=pd.read_csv('IDCJAC0010_023090_2018_Data.csv')
my_data_1=Adelaide_file_2018
my_data_1.head()
```

```python
my_data['Ade_2018']=my_data_1['Maximum temperature (Degree C)']
my_data.head()
```

```python
my_data.rename(columns={"Maximum temperature (Degree C)": "Ade_2008",}, inplace=True)
my_data.head()
```

```python
my_data_Bri_2008=pd.read_csv('IDCJAC0010_040842_2008_Data.csv')
my_data_Bri_2018=pd.read_csv('IDCJAC0010_040842_2018_Data.csv')
my_data['Bri_2008']=my_data_Bri_2008['Maximum temperature (Degree C)']
my_data['Bri_2018']=my_data_Bri_2018['Maximum temperature (Degree C)']
my_data.head()
```

```python
my_data_Can_2008=pd.read_csv('IDCJAC0010_070339_2008_Data.csv')
my_data_Can_2018=pd.read_csv('IDCJAC0010_070339_2018_Data.csv')
my_data['Can_2008']=my_data_Can_2008['Maximum temperature (Degree C)']
my_data['Can_2018']=my_data_Can_2018['Maximum temperature (Degree C)']
my_data.head()
```

```python
my_data_Dar_2008=pd.read_csv('IDCJAC0010_014015_2008_Data.csv')
my_data_Dar_2018=pd.read_csv('IDCJAC0010_014015_2018_Data.csv')
my_data['Dar_2008']=my_data_Dar_2008['Maximum temperature (Degree C)']
my_data['Dar_2018']=my_data_Dar_2018['Maximum temperature (Degree C)']
my_data.head()
```

```python
my_data_Hob_2008=pd.read_csv('IDCJAC0010_094029_2008_Data.csv')
my_data_Hob_2018=pd.read_csv('IDCJAC0010_094029_2018_Data.csv')
my_data['Hob_2008']=my_data_Hob_2008['Maximum temperature (Degree C)']
my_data['Hob_2018']=my_data_Hob_2018['Maximum temperature (Degree C)']
my_data.head()
```

```python
my_data_Mel_2008=pd.read_csv('IDCJAC0010_086038_2008_Data.csv')
my_data_Mel_2018=pd.read_csv('IDCJAC0010_086038_2018_Data.csv')
my_data['Mel_2008']=my_data_Mel_2008['Maximum temperature (Degree C)']
my_data['Mel_2018']=my_data_Mel_2018['Maximum temperature (Degree C)']
my_data.head()
```

```python
my_data_Per_2008=pd.read_csv('IDCJAC0010_009225_2008_Data.csv')
my_data_Per_2018=pd.read_csv('IDCJAC0010_009225_2018_Data.csv')
my_data['Per_2008']=my_data_Per_2008['Maximum temperature (Degree C)']
my_data['Per_2018']=my_data_Per_2018['Maximum temperature (Degree C)']
my_data.head()
```

```python
my_data_Syd_2008=pd.read_csv('IDCJAC0010_066062_2008_Data.csv')
my_data_Syd_2018=pd.read_csv('IDCJAC0010_066062_2018_Data.csv')
my_data['Syd_2008']=my_data_Syd_2008['Maximum temperature (Degree C)']
my_data['Syd_2018']=my_data_Syd_2018['Maximum temperature (Degree C)']
my_data.head()
```

```python
color = {'boxes': 'Pink', 'whiskers': 'Orange',
   ....:          'medians': 'Red', 'caps': 'Red'}
my_data_2008=my_data[['Ade_2008', 
                     'Bri_2008','Can_2008', 'Dar_2008', 'Hob_2008', 'Mel_2008', 'Per_2008', 'Syd_2008']]
my_data_2008.plot.box(color=color, sym='r+',vert=True,figsize=(15,8))
plt.title('THE MAXIMUM TEMPERATURE IN 2008',fontsize=20)
```

```python
color = {'boxes': 'Pink', 'whiskers': 'Orange',
   ....:          'medians': 'Red', 'caps': 'Red'}
my_data_2018=my_data[['Ade_2018', 
                     'Bri_2018','Can_2018', 'Dar_2018', 'Hob_2018', 'Mel_2018', 'Per_2018', 'Syd_2018']]
my_data_2018.plot.box(color=color, sym='r+',vert=True,figsize=(15,8))
plt.title('THE MAXIMUM TEMPERATURE IN 2018',fontsize=20)
```

```python
my_data['Change_Adelaide']=my_data.mean()['Ade_2018']-my_data.mean()['Ade_2008']
my_data['Change_Brisbane']=my_data.mean()['Bri_2018']-my_data.mean()['Bri_2008']
my_data['Change_Canberra']=my_data.mean()['Can_2018']-my_data.mean()['Can_2008']
my_data['Change_Darwin']=my_data.mean()['Dar_2018']-my_data.mean()['Dar_2008']
my_data['Change_Hobart']=my_data.mean()['Hob_2018']-my_data.mean()['Hob_2008']
my_data['Change_Melbourne']=my_data.mean()['Mel_2018']-my_data.mean()['Mel_2008']
my_data['Change_Perth']=my_data.mean()['Per_2018']-my_data.mean()['Per_2008']
my_data['Change_Sydney']=my_data.mean()['Syd_2018']-my_data.mean()['Syd_2008']
my_data_change=my_data[['Change_Adelaide','Change_Brisbane','Change_Canberra','Change_Darwin',
                       'Change_Hobart','Change_Melbourne','Change_Perth','Change_Sydney']]
my_data_change.iloc[:1].plot(kind='bar',figsize=(15,8),fontsize=15,);
plt.legend(fontsize = 14)
plt.title('THE MAXIMUM TEMPERATURE CHANGE FROM 2008 TO 2018',fontsize=20)
```

```python
my_data['ChangeRate_Adelaide']=(my_data.mean()['Ade_2018']-my_data.mean()['Ade_2008'])/my_data.mean()['Ade_2008']
my_data['ChangeRate_Brisbane']=(my_data.mean()['Bri_2018']-my_data.mean()['Bri_2008'])/my_data.mean()['Bri_2008']
my_data['ChangeRate_Canberra']=(my_data.mean()['Can_2018']-my_data.mean()['Can_2008'])/my_data.mean()['Can_2008']
my_data['ChangeRate_Darwin']=(my_data.mean()['Dar_2018']-my_data.mean()['Dar_2008'])/my_data.mean()['Dar_2008']
my_data['ChangeRate_Hobart']=(my_data.mean()['Hob_2018']-my_data.mean()['Hob_2008'])/my_data.mean()['Hob_2008']
my_data['ChangeRate_Melbourne']=(my_data.mean()['Mel_2018']-my_data.mean()['Mel_2008'])/my_data.mean()['Mel_2008']
my_data['ChangeRate_Perth']=(my_data.mean()['Per_2018']-my_data.mean()['Per_2008'])/my_data.mean()['Per_2008']
my_data['ChangeRate_Sydney']=(my_data.mean()['Syd_2018']-my_data.mean()['Syd_2008'])/my_data.mean()['Syd_2008']
my_data_change=my_data[['ChangeRate_Adelaide','ChangeRate_Brisbane','ChangeRate_Canberra','ChangeRate_Darwin',
                       'ChangeRate_Hobart','ChangeRate_Melbourne','ChangeRate_Perth','ChangeRate_Sydney']]
my_data_change.iloc[:1].plot(kind='bar',figsize=(15,8),fontsize=15,);
plt.legend(fontsize = 14)
plt.title('THE MAXIMUM TEMPERATURE CHANGE RATE FROM 2008 TO 2018',fontsize=20)
```
